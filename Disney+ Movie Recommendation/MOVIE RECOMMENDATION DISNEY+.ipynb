## SEARCHING FOR THE MOVIE OR SHOW BASED ON GENRE

```python
def genre_type_movies(genre):
    condition= movies.genres.apply(lambda x: genre in x)
    Filtered = movies[condition]
    best5M = Filtered[["title","imdb_score","genres"]].sort_values(by="imdb_score",ascending=False)
    return best5M.head(5)
```

```python
genre = input()
genre_type_movies(genre)
```

```python
def genre_type_shows(genre):
    condition= shows.genres.apply(lambda x: genre in x)
    Filtered = shows[condition]
    best5M = Filtered[["title","imdb_score","genres"]].sort_values(by="imdb_score",ascending=False)
    
    return best5M.reset_index(drop=True).head(5)
```

```python
genre = input()
genre_type_movies(genre)
```
